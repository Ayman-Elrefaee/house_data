```python
Data_Set = pd.read_csv('/content/drive/MyDrive/SQL_Pthon/kc_house_data_NaN.csv')
```

```python
Data_Set.head(100)
```

```python
# Drop the columns "id" and "Unnamed: 0" from axis 1 using the method drop(), then use the method describe() to obtain a statistical summary of the data. Make sure the inplace parameter is set to True.

Data_Set.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
Data_Set.describe()
```

```python
Data_Set.isna().sum()
```

```python
mean=Data_Set['bedrooms'].mean()
Data_Set['bedrooms'].replace(np.nan,mean, inplace=True)
```

```python
mean = Data_Set['bathrooms'].mean()
Data_Set['bathrooms'].replace(np.nan, mean, inplace=True)
```

```python
# prompt: set data type for bedrooms , floors and bathrooms sa int

Data_Set[['bedrooms', 'bathrooms', 'floors']] = Data_Set[['bedrooms', 'bathrooms','floors']].astype(int)
```

```python
# date column as date type
Data_Set['date'] = pd.to_datetime(Data_Set['date'])
```

```python
Data_Set.info()
```

```python
#We can use the Pandas method corr() to find the feature other than price that is most correlated with price.
df_numeric = Data_Set.select_dtypes(include=[np.number])
df_numeric.corr()['price'].sort_values()
```

```python
# prompt: Drop columns: lat , long ,zipcode

Data_Set.drop(['lat', 'long', 'zipcode'], axis=1, inplace=True)
```

```python
Data_Set.head(10)
```

```python
#  Drop duplicated rows
Data_Set.drop_duplicates(inplace=True)
```

```python
#unique value in water front column

Data_Set['waterfront'].unique()
```

```python
# prompt: if waterfront = 0 repalce it with 'no waterfront'
# else replace with has 'waterfront'

Data_Set['waterfront'] = Data_Set['waterfront'].apply(lambda x: 'no waterfront' if x == 0 else 'has waterfront')
print(Data_Set['waterfront'].value_counts())
```

```python
Data_Set.head()
```

```python
# prompt: change unite sqft(square feet) to square meter for all columns start with sqft and rename this columns

# Define the conversion factor from square feet to square meters
sqft_to_sqm = 0.092903

# Iterate through columns that start with 'sqft'
for col in Data_Set.columns:
  if col.startswith('sqft'):
    # Convert sqft values to square meters
    Data_Set[col] = Data_Set[col] * sqft_to_sqm
    # Rename the column to indicate it's now in square meters
    new_col_name = col.replace('sqft', 'sqm')
    Data_Set.rename(columns={col: new_col_name}, inplace=True)
```

```python
# prompt:  columns starts with('sqm') change type to int

# Iterate through columns that start with 'sqm' and change their type to int
for col in Data_Set.columns:
  if col.startswith('sqm'):
    Data_Set[col] = Data_Set[col].astype(int)

Data_Set.info()
```

```python
Data_Set.head()
```

```python
# prompt: Creat ID column Start by 1 and its datatype is object

# Create an 'ID' column starting from 1 and with object datatype
Data_Set.insert(0, 'ID', range(1, 1 + len(Data_Set)))
Data_Set['ID'] = Data_Set['ID'].astype(object)

# Display the first few rows to verify
Data_Set.head()

# Display the info to verify the datatype
Data_Set.info()
```

```python
# prompt: import this Data_Set as CSV file to MyDrive

Data_Set.to_csv('/content/drive/MyDrive/processed_kc_house_data.csv', index=False)
```
